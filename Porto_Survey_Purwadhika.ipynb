# Analisis untuk Pembuatan Strategi Inklusivitas 

## 1. Sumber Informasi Utama

```python
# buat visualisasi data sumber informasi
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Informasi_Purwadhika')
plt.title('Distribusi Sumber Informasi')
```

```python
from scipy.stats import chi2_contingency

# Hitung frekuensi sumber informasi
sumber_info = df['Informasi_Purwadhika'].value_counts()

# Buat tabel kontingensi (jika dibandingkan dengan variabel lain, misal kelas)
tabel_kontingensi = pd.crosstab(df['Informasi_Purwadhika'], df['kelas'])
chi2, p_value, _, _ = chi2_contingency(tabel_kontingensi)

print("Frekuensi Sumber Informasi:")
print(sumber_info)
print(f"\nChi-square: {chi2:.2f}, P-value: {p_value:.4f}\n")

H0 = 'Tidak ada hubungan antara sumber informasi dengan pemilihan kelas'
Ha = 'Terdapat hubungan antara sumber informasi dengan pemilihan kelas'

if p_value < 0.05:
    print(f'Tolak H0: {Ha}')
else:
    print(f'Terima H0: {H0}')
```

## 2. Asal Daerah

```python
# buat visualisasi data asal daerah
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Asal_Daerah')
plt.title('Distribusi Asal Daerah')
```

```python
# cek pengaruh asal daerah terhadap pemilihan kelas
asal_daerah = df['Asal_Daerah'].value_counts()

# Buat tabel kontingensi (jika dibandingkan dengan variabel lain, misal kelas)
tabel_kontingensi = pd.crosstab(df['Asal_Daerah'], df['kelas'])
chi2, p_value, _, _ = chi2_contingency(tabel_kontingensi)

print("Frekuensi Asal Daerah:")
print(asal_daerah)
print(f"\nChi-square: {chi2:.2f}, P-value: {p_value:.4f}\n")

H0 = 'Tidak ada hubungan antara asal daerah dengan kelas'
Ha = 'Terdapat hubungan antara asal daerah dengan kelas'

if p_value < 0.05:
    print(f'Tolak H0: {Ha}')
else:
    print(f'Terima H0: {H0}')
```

```python
# cek pengaruh daerah asal dengan tingkat kemampuan
contingency_table = pd.crosstab(df['Asal_Daerah'], df['Tingkat_Kemampuan'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Test Results (Origin vs. Skill Level):")
print(f"Chi2 statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")

alpha = 0.05 
if p < alpha:
    print("H0 ditolak: Ada hubungan signifikan antara asal daerah dan tingkat kemampuan siswa.")
else:
    print("H0 diterima: Tidak ada hubungan signifikan antara asal daerah dan tingkat kemampuan siswa.")

# custom visualisasi heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt="d", cmap="Blues_r", cbar=True)
plt.title('Hubungan antara Asal Daerah dan Tingkat Kemampuan Siswa')
plt.xlabel('Tingkat Kemampuan')
plt.ylabel('Asal Daerah')
plt.show()
```

## 3. Perbedaan Pengeluaran Bulanan

```python
#buat visualisasi rerata data pengeluaran bulanan dengan diagram batang
plt.figure(figsize=(10, 6))
sns.barplot(x='kelas', y='Pengeluaran_Bulanan', data=df)
plt.title('Rerata Pengeluaran Bulanan Berdasarkan Kelas')
```

```python
# analisis perbedaan pengeluaran bulanan dari murid kelas WD dan DM
from scipy.stats import mannwhitneyu

wd = df[df['kelas']=='WD']['Pengeluaran_Bulanan']
dm = df[df['kelas']=='DM']['Pengeluaran_Bulanan']

# uji t-test independent
from scipy import stats
stat, p_value = mannwhitneyu(wd, dm)

print(f"Mann-Whitney U: {stat:.2f}, P-value: {p_value:.4f}\n")

H0 = 'Tidak ada perbedaan signifikan antara pengeluaran bulanan murid kelas WD dan DM'
Ha = 'Terdapat perbedaan signifikan antara pengeluaran bulanan murid kelas WD dan DM'

if p_value < 0.05:
    print(f'Tolak H0: {Ha}')
else:
    print(f'Terima H0: {H0}')
```

## 4. Pendidikan dan Tingkat Kemampuan

```python
# cek apakah tingkat pendidikan berpengaruh terhadap tingkat kemampuan
crosstab = pd.crosstab(df['Pendidikan'], df['Tingkat_Kemampuan'])

plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="Blues_r", linewidths=.5)  #
plt.title('Tingkat Kemampuan vs Pendidikan')
plt.xlabel('Tingkat Kemampuan')
plt.ylabel('Pendidikan')
plt.show()

# visualisasi data pendidikan dan tingkat kemampuan
crosstab_percentage = crosstab.div(crosstab.sum(axis=1), axis=0) * 100
crosstab_percentage.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis') 
plt.title('Proporsi Tingkat Kemampuan berdasarkan Pendidikan')
plt.xlabel('Pendidikan')
plt.ylabel('Persentase')
plt.legend(title='Tingkat Kemampuan')
plt.show()
```

```python
from scipy.stats import chi2_contingency

tabel_kontingensi = pd.crosstab(df['Pendidikan'], df['Tingkat_Kemampuan'])
chi2, p_value, _, _ = chi2_contingency(tabel_kontingensi)

print("Tabel Kontingensi Pendidikan vs Tingkat Kemampuan:")
print(tabel_kontingensi)
print(f"\nChi-square: {chi2:.2f}, P-value: {p_value:.4f}\n")

H0 = 'Tidak ada hubungan antara tingkat pendidikan dengan tingkat kemampuan'
Ha = 'Terdapat hubungan antara tingkat pendidikan dengan tingkat kemampuan'

if p_value < 0.05:
    print(f'Tolak H0: {Ha}')
else:
    print(f'Terima H0: {H0}')
```

## 6. Umur Peserta

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Umur'], kde=True, bins=range(df['Umur'].min(), df['Umur'].max() + 2, 1)) 
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(range(df['Umur'].min(), df['Umur'].max() + 1)) 
plt.show()
```

```python
from scipy.stats import kruskal

dm = df[df['kelas'] == 'DM']['Umur']
wd = df[df['kelas'] == 'WD']['Umur']
stat, p_value = kruskal(dm, wd)

print(f"Kruskal-Wallis: {stat:.2f}, P-value: {p_value:.4f}\n")

H0 = 'Tidak ada perbedaan signifikan antara umur murid kelas DM dan WD'
Ha = 'Terdapat perbedaan signifikan antara umur murid kelas DM dan WD'

if p_value < 0.05:
    print(f'Tolak H0: {Ha}')
    print(f'Umur rata-rata murid kelas DM: {dm.mean():.2f}')
    print(f'Umur rata-rata murid kelas WD: {wd.mean():.2f}')
    print(f'Perbedaan rata-rata: {dm.mean() - wd.mean():.2f}')
else:
    print(f'Terima H0: {H0}')
    print(f'Umur rata-rata murid kelas DM: {dm.mean():.2f}')
    print(f'Umur rata-rata murid kelas WD: {wd.mean():.2f}')
    print(f'Perbedaan rata-rata: {dm.mean() - wd.mean():.2f}')
```

```python
# cek apakah umur berpengaruh signifikan terhadap pemilihan kelas

from scipy.stats import kruskal

dm = df[df['kelas'] == 'DM']['Umur']
wd = df[df['kelas'] == 'WD']['Umur']
stat, p_value = kruskal(dm, wd)

print(f"Kruskal-Wallis: {stat:.2f}, P-value: {p_value:.4f}\n")

H0 = 'Tidak ada perbedaan signifikan antara umur murid kelas DM dan WD'
Ha = 'Terdapat perbedaan signifikan antara umur murid kelas DM dan WD'

if p_value < 0.05:
    print(f'Tolak H0: {Ha}')
    print(f'Umur rata-rata murid kelas DM: {dm.mean():.2f}')
    print(f'Umur rata-rata murid kelas WD: {wd.mean():.2f}')
    print(f'Perbedaan rata-rata: {dm.mean() - wd.mean():.2f}')
else:
    print(f'Terima H0: {H0}')
    print(f'Umur rata-rata murid kelas DM: {dm.mean():.2f}')
    print(f'Umur rata-rata murid kelas WD: {wd.mean():.2f}')
    print(f'Perbedaan rata-rata: {dm.mean() - wd.mean():.2f}')
```

```python
# cek apakah umur berpengaruh signifikan terhadap tingkat kemampuan

from scipy.stats import kruskal

skill_levels = []
for skill in df['Tingkat_Kemampuan'].unique():  
    skill_levels.append(df[df['Tingkat_Kemampuan'] == skill]['Umur'])

if len(skill_levels) >= 2: 
    statistic, p_value = kruskal(*skill_levels)

    print("Kruskal-Wallis Test Results:")
    print(f"Statistic: {statistic}")
    print(f"P-value: {p_value}")

    alpha = 0.05  
    if p_value < alpha:
        print("H0 ditolak: Ada perbedaan signifikan antara umur dan tingkat kemampuan siswa.")
    else:
        print("H0 diterima: Tidak ada perbedaan signifikan antara umur dan tingkat kemampuan siswa.")

else:
    print("Tidak cukup tingkat kemampuan yang berbeda untuk melakukan uji statistik")

# visualisasi boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Tingkat_Kemampuan', y='Umur', data=df)
plt.title('Pengaruh Umur terhadap Tingkat Kemampuan Siswa')
plt.xlabel('Tingkat Kemampuan')
plt.ylabel('Umur')
plt.show()
```
